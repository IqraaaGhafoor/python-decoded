# Variables and Types
Variables are the symbolic names that references or print ato an object in memory.  They are used to store and manipulate data in a program.
These are basically like a container where you put data.

Think of a variable as a labeled box where you can place data and retrieve whhen needed. 
straightforward in python 
assign a value to a variable name using the assignment operator (=)
# Rules for variable naming 
1. Case sensitive
2. Start with a letter or underscore
   
 valid: _value, data1

 invalid: 1data, #score

1. Followed by leters , digits, or underscore 
   
 valid: value_1, data_set
 
 invalid: data-set, value-1 (hyphen are not allowed.)

# Variables naming conventions
1. Use lowercase letters
2. Avoid reserved keywords
3. Meaningful names

## Math Oprations

###  + means add
###  - means subtract
###  * means multiply
###  / means divide or division
###  ** means power
###  % means remainder
###  // means Floor Division (whole number only)
###  += means keep adding to something
###  -= means keep subtracting
###  *= means multiply what you already have
###  /= means divide what you already have
###  //= means divide but keep whole number
###  %= means find leftover after dividing
###  **= means raise to a power repeatedly

## Built-in Math Functions

### abs(-50)            # 50     removes minus sign
### round(3.7)          # 4      rounds to nearest whole number
### round(3.14159, 2)   # 3.14   rounds to 2 decimal places
### max(10, 50, 30)     # 50     biggest number
### min(10, 50, 30)     # 10     smallest number
### sum([10, 20, 30])   # 60     adds all numbers
### pow(2, 8)           # 256    same as 2 ** 8

## Comparison Operations

### print(a == b)    equal
### print(a != b)    NOT equal
### print(a > b)     greater than
### print(a < b)     less than
### print(a >= 10)   greater than OR equal
### print(a <= 5)    less than OR equal


```python
a = 10
b = 5

sum_result = a + b
difference = a - b
product = a * b
quotient = a / b
remainder = a % b
power = a ** b

print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
print("Remainder:", remainder)
print("Power:", power)
```
