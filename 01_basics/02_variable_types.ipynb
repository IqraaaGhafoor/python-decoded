# 1. Integer (int)

```python
age = 10 
print(type(age))
```

```python
# Basic arithmetic operations with integers
a = 555
b = 99

sum_result = a + b
difference = a - b
product = a * b
quotient = a // b  # Integer division (no decimal part)
remainder = a % b  # Modulus (remainder of division)
power = a ** 2
power = b ** 3

print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
print("Remainder:", remainder)
print("Power:", power)
print("Power:", power)

# Type
print(type(sum_result))
print(type(difference))
print(type(product))
print(type(quotient))
print(type(remainder))
print(type(power))
```

```python
a = 220
b = 440

if a > b:
    print(a, "is greater than", b)
elif a < b:
    print(a, "is less than", b)
else:
    print(a, "is equal to", b)
```

```python
numbers = [33, 89, 111, 44,107]

max_value = max(numbers)
min_value = min(numbers)

print("Maximum value:", max_value)
print("Minimum value:", min_value)

print(type(max_value))
print(type(min_value))
```
