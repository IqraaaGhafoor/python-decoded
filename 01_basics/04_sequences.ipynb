# Sequnce in Python

Sequences are a type of data structure that allows you to store multiple items. 

The main types of sequences in Python are:

### Lists (list) 

Ordered, mutable (can be changed), and can contain different types of elements.

### Tuples (tuple)

Ordered, immutable (cannot be changed), and can also contain different types of elements.

### Ranges (range) 

Used to represent an immutable sequence of numbers.

### Strings (str)

A sequence of characters.

```python
# length of a sequence  : we can use the len() function.

my_list = [1, 2, 3, 4,5]
print(len(my_list))
```

```python
fruits = ["apple", "banana", "cherry", "apricot", "watermelon"]
print(len(fruits))
```

```python
# Membership Test

fruits = ["apple", "banana", "cherry", "apricot", "watermelon"]
print("apple" in fruits)
print("cherry" not in fruits)
```

```python
# Concatenation 

list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
list1 + list2
```

```python
# Repetition

my_list = [2,3,4,5,6,7,8]
my_list*4
```
