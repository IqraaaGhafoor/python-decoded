# Mutable Elments

Mutable means these are changeable.

```python
list = ["Apple", "Mango", "Cherry", "Bnanana", "Melon"]
```

```python
list[3] = "Grapes"
```

```python
print(list)
```

```python
numbers = [1, 2, 3]

numbers.append(4)   #append means adding a single item.
numbers[0] = 10

print(numbers)
```

```python
list = ["Icecream", "Jam"]

list.append("Honey")
list.remove("Jam")

print(list)
```

```python
colors = ["red", "green", "blue"]

colors[1] = "yellow"    #change item

colors.append("black")   #add item
colors.remove("red")     #remove item

print(colors)
```

```python
student = {
    "name": "Ahmad",
    "age": 20
}

student["age"] = 21  #modify/changing the value

student["city"] = "London"   #adding new key

print(student)
```
