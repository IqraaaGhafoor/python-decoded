# Type casting in Python

Change one data type to another data type

### Types of Type casting

#### Implicit Type casting:

means python is doing. Like automative type casting

#### Explicit Type casting 

menas you are doing like changing manually one data type to another data type

## 1. Casting int to float and float to int

```python
num_int = 256
num_float = float(num_int)
print(num_float)
print(type(num_float))
```

```python
num_float = 4.98
num_int = int(num_float)
print(num_int)
print(type(num_int))
```

```python
age_int = 31
age_float = float(age_int)
print(age_float)
print(type(age_int))
print(type(age_float))
```

## 2. casting float or int to str

```python
num_float = 34.7898
num_str = str(num_float)
print(num_str)
print(type(num_str))
```

```python
num_int = 237
num_str = str(num_int)
print(num_str)
print(type(num_str))
```

## 3. Casting str to int and float

```python
num_str = "123"
num_float = float(num_str)
print(type(num_str))
```

```python
num_int = 1222222
num_float = float(num_int)
print(num_float)
print(type(num_float))
```

## 4. casting list to tuple and tuple to list

```python
# List is mutable, meaning we can modify it (add/remove items).

num_list = [1, 2, 3, 4, 5, 6]
num_tuple = tuple(num_list)
print(num_tuple)
print(type(num_tuple))
```

```python
# Tuple is immutable, meaning once created, its values cannot be changed.

num_tuple = (2, 3, 4, 6, 7, 8)
num_list = list(num_tuple)
print(num_list)
print(type(num_list))
```

```python
my_list = [1, 2, "apple", "dates"]
my_tuple = tuple(my_list)
print(my_tuple)
```

## 5. Casting list to set and set to list

```python
# Lists are ordered and can contain duplicates.

num_list = [1, 2, 3, 6, 9]
num_set = set(num_list)
print(num_set)
print(type(num_set))
```

```python
# Sets are unordered and do not allow duplicates.

num_set = {45, 67, "apple"}
num_list = list(num_set)
print(num_list)
print(type(num_list))
```

## 6. too boolean

```python
num = 10
num_bool = bool(num)
print(num_bool)
print(type(num_bool))
```

```python
num = []
num_bool = bool(num)
print(num_bool)
print(type(num_bool))
```

## Dynamic Type casting

### Implicit Typecasting

```python
# Implicit typecasting
a = 10
b = 12.5
#adding int and float
c = a + b   #result in a float
print(c)
print(type(c))
```

### Explicit Type casting

```python
num_str = "234"    # int(), float(), and str() etc. 
num_int = int(num_str)   #manually converting string to int.
print(num_int)
print(type(num_int))
```

# Python Typecasting & Operator Rules 

 Python behavior depends on **data types**, not values or intention  
 Operators act based on **type combinations**  
 '+' works only for **same types** (e.g., number+number, string+string)  
 '*' supports **repetition** when used with sequences and integers  
 '-', /, //, % work **only with numeric types**  
 Typecasting creates a **new object**, does not modify original    
 **Booleans behave like integers** ('True = 1', 'False = 0')  
 **Sequences follow rules**: same type → combine, int → repeat  
 **Truthiness rule**: empty = False, non-empty = True  

```python
x = 5
y = 2.0
print(type(x + y))
```

```python
x = "5"
y = 2

result = x * y + str(int(x) + y)
print(result)
```
